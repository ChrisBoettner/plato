```python
def get_kinematic_characteristics(Z, R, component, dataframe):
    component_map = {
        "thin_disk": "_D",
        "thick_disk": "_TD",
        "halo": "_H",
    }
    if component not in component_map.keys():
        raise ValueError(
            f"Component must be one of {component_map.keys()}, got {component}"
        )

    if np.abs(Z) < dataframe["Z"].min():
        Z = dataframe["Z"].min()
    if R < dataframe["R"].min():
        R = dataframe["R"].min()

    matching_row = dataframe[
        (dataframe["Z"] <= np.abs(Z)) & (dataframe["R"] <= R)
    ].iloc[-1]
    # return columns that end in component_map[component]
    matching_row = matching_row[
        [col for col in matching_row.index if col.endswith(component_map[component])]
    ]
    # rename columns to remove the component_map
    matching_row = matching_row.rename(
        {col: col.replace(component_map[component], "") for col in matching_row.index}
    )
    return matching_row
```

```python
get_kinematic_characteristics([0.56, 0.2], [8.2, 1], "thin_disk", data)
```
